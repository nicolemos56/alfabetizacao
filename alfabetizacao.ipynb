```python
from google.colab import files
from google.cloud import vision
from google.colab import auth; auth.authenticate_user()
import io



# Autenticação (não precisa de API Key no Colab)
# O Colab se encarrega da autenticação

# Instalar a biblioteca do Google Cloud Vision (caso não esteja instalada)
!pip install google-cloud-vision
from google.colab import auth
auth.authenticate_user()

def detectar_texto(image_content):
  """Detecta texto em uma imagem."""
  client = vision.ImageAnnotatorClient()
  image = vision.Image(content=image_content)
  response = client.text_detection(image=image)

  if response.text_annotations:
      texto_detectado = response.text_annotations[0].description
      return texto_detectado
  else:
      return "Nenhum texto detectado"

      # Carregar a imagem do computador
uploaded = files.upload()

# Obter o conteúdo da imagem
file_name = next(iter(uploaded))
image_content = uploaded[file_name]

# Detectar texto na imagem
texto = detectar_texto(image_content)

print(f"Texto detectado: {texto}")
```
